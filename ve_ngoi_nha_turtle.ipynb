```python
import turtle
screen = turtle.Screen()
screen.setup(1000,1000)
screen.title("Ngoi nha mo uoc")
t = turtle.Turtle()

#doi mau man hinh

#nhac but len di den diem (250,250)
t.penup()
t.goto(250,250)

#ve mat troi
#doi mau but thanh den, dat but xuong
t.pendown()
t.pencolor("black")
#ve tia sang
t.fd(60)
t.bk(120)
t.fd(60)
t.lt(45)

t.fd(80)
t.bk(160)
t.fd(80)
t.lt(45)

t.fd(60)
t.bk(120)
t.fd(60)
t.lt(45)

t.fd(80)
t.bk(160)
t.fd(50)
t.rt(90)

#to mau cho hinh tron mau do
t.fillcolor("yellow")
t.begin_fill()
#dat ban kinh hinh tron 30
t.circle(30)
t.end_fill()


#ve cay
#nhac but len di den diem (400,100)
t.penup()
t.goto(400,100)
#doi mau but thanh xanh, dat but xuong
t.pendown()

#to mau cho hinh tron mau xanh
t.fillcolor("green")
t.begin_fill()
t.goto(420,70)
t.goto(380,70)
t.goto(400,100)
t.penup()
t.goto(400,70)
t.pendown()
t.goto(420,40)
t.goto(380,40)
t.goto(400,70)
t.penup()
t.goto(400,40)
t.pendown()
t.goto(420,10)
t.goto(380,10)
t.goto(400,40)
t.end_fill()

#to mau goc cay
t.penup()
t.goto(405,10)
t.pendown()
t.fillcolor("brown")
t.begin_fill()
t.rt(135)
t.fd(40)
t.rt(90)
t.fd(10)
t.rt(90)
t.fd(40)
t.rt(90)
t.fd(10)
t.end_fill()

#ve nha
#nhac but len, di ve diem goc
t.penup()
t.home()

#to mau tuong truoc
t.pendown()

t.fillcolor("blue")
t.begin_fill()
t.rt(90)
t.fd(100)
t.rt(90)
t.fd(60)
t.rt(90)
t.fd(100)
t.rt(90)
t.fd(60)
t.end_fill()

#to mai tuong truoc

t.fillcolor("violet")
t.begin_fill()
t.bk(60)
t.goto(-30,30)
t.home()
t.end_fill()

#to mau tuong bên

t.fillcolor("yellow")
t.begin_fill()
t.rt(90)
t.fd(100)
t.goto(100,-85)
t.goto(100,15)
t.home()
t.end_fill()

#to mai tuong bên

t.fillcolor("orange")
t.begin_fill()
t.goto(-30,30)
t.goto(70,45)
t.goto(100,15)
t.home()
t.end_fill()

#ve cua trc
t.penup()
t.goto(-20,-30)
t.pendown()

t.fillcolor("brown")
t.begin_fill()
t.goto(-20,-100)
t.goto(-40,-100)
t.goto(-40,-30)
t.goto(-20,-30)
t.end_fill()

#ve cua xo
t.penup()
t.goto(40,-60)
t.pendown()

t.fillcolor("brown")
t.begin_fill()
t.goto(40,-40)
t.goto(80,-34)
t.goto(80,-54)
t.goto(40,-60)
t.end_fill()





turtle.done()
```
